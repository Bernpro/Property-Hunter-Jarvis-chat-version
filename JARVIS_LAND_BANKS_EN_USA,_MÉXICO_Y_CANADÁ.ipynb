```python
# ============================================
# JARVIS CON LAND BANKS - USA, MÉXICO Y CANADÁ
# Versión sin instalaciones - Funciona en Google Colab
# ============================================

import datetime
import random
import urllib.parse
from IPython.display import display, HTML, clear_output
import ipywidgets as widgets

clear_output(wait=True)

print("="*65)
print("🌎 JARVIS - LAND BANKS EN USA, MÉXICO Y CANADÁ")
print("="*65)
print("✅ Busca land banks y propiedades gubernamentales en:")
print("   🇺🇸 Estados Unidos - Tucson, Phoenix, Las Vegas, Los Angeles")
print("   🇲🇽 México - CDMX, Puebla, Aguascalientes")
print("   🇨🇦 Canadá - Federal (Canada Public Land Bank), Ontario, Regina")
print("="*65)

# ============================================
# BASE DE DATOS DE LAND BANKS
# ============================================

LAND_BANKS = {
    # ===== ESTADOS UNIDOS =====
    'tucson': {
        'pais': '🇺🇸 Estados Unidos',
        'nombre': 'Southern Arizona Land Trust (SALT)',
        'ubicacion': 'Tucson, Arizona',
        'county': 'Pima County',
        'descripcion': 'Organización sin fines de lucro que revitaliza áreas afectadas por la recesión, con 128 propiedades en alquiler y múltiples proyectos en desarrollo.',
        'sitio_web': 'https://www.saltaz.org',
        'propiedades_url': 'https://www.saltaz.org/properties',
        'programas': [
            '🏠 Propiedades en alquiler - 128 unidades (20% para familias de bajos ingresos)',
            '🏗️ Belnor Vista - Desarrollo de viviendas',
            '🏘️ Bentley on the Loop - 3 nuevas casas',
            '🔨 Emery Park Place - 4 dúplex',
            '📐 Pinal Vista Place - 6 parcelas para viviendas'
        ],
        'contacto': '3044 N Alvernon Way, Tucson, AZ 85712'
    },

    'phoenix': {
        'pais': '🇺🇸 Estados Unidos',
        'nombre': 'Maricopa County Land Bank / City of Phoenix',
        'ubicacion': 'Phoenix, Arizona',
        'county': 'Maricopa County',
        'descripcion': 'Programas de estabilización de vecindarios y propiedades en ejecución fiscal del condado de Maricopa.',
        'sitio_web': 'https://www.maricopa.gov/5647/Land-Bank',
        'propiedades_url': 'https://www.maricopa.gov/5651/Available-Properties',
        'programas': [
            '🏠 Neighborhood Stabilization Program - Viviendas recuperadas',
            '📋 Tax Lien Sales - Propiedades con deuda fiscal',
            '🔨 Land Reutilization - Terrenos para reurbanización',
            '🏘️ Affordable Housing Program - Vivienda asequible'
        ],
        'contacto': '301 W Jefferson St, Phoenix, AZ 85003'
    },

    'las vegas': {
        'pais': '🇺🇸 Estados Unidos',
        'nombre': 'Clark County Land Bank (Land Reutilization Corporation)',
        'ubicacion': 'Las Vegas, Nevada',
        'county': 'Clark County',
        'descripcion': 'Programa establecido para eliminar deterioro urbano y estabilizar vecindarios mediante la adquisición y reurbanización de propiedades problemáticas.',
        'sitio_web': 'https://clarkcountylandbank.org',
        'propiedades_url': 'https://clarkcountylandbank.org/available-properties',
        'programas': [
            '🏠 Side Lot Program - Terrenos vacantes adyacentes',
            '🏗️ Vacant Lot Program - Terrenos NO adyacentes para construir',
            '🏢 Commercial Property - Desarrollo económico',
            '🔨 Rehabilitation Program - Renovación con ocupación de 3 años',
            '🏚️ Demolition Program - 132 demoliciones completadas'
        ],
        'target_areas': [
            'Grand Ave South / Promise Neighborhood',
            'Highlands / Southgate Neighborhood',
            'Lagonda Redevelopment Corridor',
            'Sheridan / Kenton Corridor',
            'Western Edge'
        ],
        'contacto': '500 S Grand Central Pkwy, Las Vegas, NV 89106'
    },

    'los angeles': {
        'pais': '🇺🇸 Estados Unidos',
        'nombre': 'Los Angeles County Land Bank Pilot Program',
        'ubicacion': 'Los Angeles, California',
        'county': 'Los Angeles County',
        'descripcion': 'Programa piloto lanzado en 2022 con $50 millones para prevenir desplazamiento cerca de nuevas infraestructuras (Metro y LA River).',
        'sitio_web': 'https://ceo.lacounty.gov/land-bank',
        'propiedades_url': 'https://ceo.lacounty.gov/land-bank/available-properties',
        'programas': [
            '🏗️ Adquisición cerca de estaciones de Metro para vivienda asequible',
            '🏠 2615 W. 8th Street - $4.3M, 160 unidades residenciales + retail',
            '🚆 Propiedad cerca de A Line - 160 viviendas + usos comerciales',
            '🤝 Colaboración con LA Community Land Trust Coalition',
            '🚇 Coordinación con Metro para propiedades cerca de tránsito'
        ],
        'contacto': '500 W Temple St, Los Angeles, CA 90012'
    },

    # ===== MÉXICO =====
    'cdmx': {
        'pais': '🇲🇽 México',
        'nombre': 'Banco de Tierras - Ciudad de México',
        'ubicacion': 'Ciudad de México',
        'descripcion': 'El gobierno de la CDMX ha implementado bancos de tierras para la gestión de reservas territoriales. Un ejemplo destacado son las 55 hectáreas de la antigua refinería de Pemex en Azcapotzalco, que estuvieron sin uso por tres décadas hasta convertirse en un parque ecológico en 2011 con lago, jardines y canchas deportivas [citation:1].',
        'sitio_web': 'https://www.cdmx.gob.mx',
        'propiedades_url': 'https://www.cdmx.gob.mx/vivienda',
        'programas': [
            '🏗️ Desarrollo de unidades habitacionales nuevas',
            '🔨 Rehabilitación de unidades existentes',
            '🏞️ Mantenimiento para uso futuro',
            '🌳 Parque ecológico Azcapotzalco (55 hectáreas)',
            '🏘️ Vivienda para población de bajos ingresos'
        ],
        'contacto': 'Gobierno de la Ciudad de México'
    },

    'puebla': {
        'pais': '🇲🇽 México',
        'nombre': 'Banco de Tierras - Puebla',
        'ubicacion': 'Puebla, México',
        'descripcion': 'El gobierno estatal de Puebla, a través de sus instituciones de vivienda, ha adquirido reservas territoriales para su urbanización y colocación entre constructores de vivienda, atendiendo las necesidades de vivienda de la población de bajos ingresos [citation:1].',
        'sitio_web': 'https://www.puebla.gob.mx',
        'propiedades_url': 'https://www.puebla.gob.mx/vivienda',
        'programas': [
            '🏗️ Adquisición de reservas territoriales',
            '🏘️ Urbanización para constructores de vivienda',
            '📐 Vivienda para población de bajos ingresos',
            '🔨 Desarrollo de proyectos habitacionales'
        ],
        'contacto': 'Gobierno del Estado de Puebla'
    },

    'aguascalientes': {
        'pais': '🇲🇽 México',
        'nombre': 'Banco de Tierras - Aguascalientes',
        'ubicacion': 'Aguascalientes, México',
        'descripcion': 'El gobierno de Aguascalientes ha adquirido reservas territoriales enmarcadas en el plan de desarrollo urbano del estado, que busca ordenar el crecimiento con una visión de 30 años [citation:1].',
        'sitio_web': 'https://www.aguascalientes.gob.mx',
        'propiedades_url': 'https://www.aguascalientes.gob.mx/vivienda',
        'programas': [
            '🏗️ Reservas territoriales para vivienda',
            '📐 Plan de desarrollo urbano con visión 30 años',
            '🏘️ Vivienda para población de bajos ingresos',
            '🔨 Desarrollo de proyectos habitacionales'
        ],
        'contacto': 'Gobierno del Estado de Aguascalientes'
    },

    # ===== CANADÁ =====
    'canada federal': {
        'pais': '🇨🇦 Canadá',
        'nombre': 'Canada Public Land Bank',
        'ubicacion': 'Federal - Todo Canadá',
        'descripcion': 'Lanzado en agosto de 2024 como parte del "Public Lands for Homes Plan". Actualmente lista 90 propiedades que representan aproximadamente 42,000 unidades de vivienda. El plan busca desbloquear 250,000 unidades para 2031 [citation:3][citation:7].',
        'sitio_web': 'https://www.canada.ca/en/public-services-procurement/services/public-land-bank.html',
        'propiedades_url': 'https://www.canada.ca/en/public-services-procurement/services/public-land-bank/map.html',
        'programas': [
            '🏗️ 90 propiedades federales identificadas',
            '🏠 42,000 unidades de vivienda potenciales',
            '📐 430 hectáreas de tierra disponible',
            '💰 $500 millones para adquisición de tierras',
            '🤝 Colaboración con CMHC y Canada Lands Company'
        ],
        'contacto': 'Public Services and Procurement Canada'
    },

    'ontario': {
        'pais': '🇨🇦 Canadá',
        'nombre': 'Edwardsburgh Landbank - Ontario',
        'ubicacion': 'Edwardsburgh-Cardinal, Ontario',
        'descripcion': 'Uno de los cuatro land banks creados por el gobierno de Ontario en 1975. Consta de 32 parcelas en 3 grupos de propiedades, totalizando 10,000 acres. Incluye tierras para vivienda, desarrollo industrial, agricultura y conservación [citation:5][citation:13].',
        'sitio_web': 'https://www.edwardsburghcardinal.ca',
        'propiedades_url': 'https://www.edwardsburghcardinal.ca/landbank',
        'programas': [
            '🏠 Desarrollo de vivienda',
            '🏢 Desarrollo industrial',
            '🌾 Tierras agrícolas activas',
            '🌳 Conservación ambiental'
        ],
        'contacto': 'Township of Edwardsburgh-Cardinal'
    },

    'regina': {
        'pais': '🇨🇦 Canadá',
        'nombre': 'Regina Land Bank',
        'ubicacion': 'Regina, Saskatchewan',
        'descripcion': 'Establecido en 2024 con fondos del Federal Housing Accelerator Fund. Actualmente tiene 16 propiedades para desarrollo futuro de vivienda. También estableció fideicomisos de tierras en los vecindarios North Central y Heritage [citation:2].',
        'sitio_web': 'https://www.regina.ca',
        'propiedades_url': 'https://www.regina.ca/housing/land-bank',
        'programas': [
            '🏠 16 propiedades para desarrollo de vivienda',
            '🏘️ North Central Revitalization Initiative',
            '🤝 Fideicomisos de tierras comunitarios',
            '🏗️ Vivienda por debajo del precio de mercado'
        ],
        'contacto': 'City of Regina'
    }
}

# ============================================
# FUNCIONES DE BÚSQUEDA
# ============================================

def buscar_land_bank(consulta):
    """Busca land bank por ciudad o país"""
    consulta_lower = consulta.lower()

    sinonimos = {
        # USA
        'tucson': ['tucson', 'southern arizona', 'salt', 'pima county'],
        'phoenix': ['phoenix', 'maricopa', 'maricopa county'],
        'las vegas': ['las vegas', 'vegas', 'clark county', 'clark'],
        'los angeles': ['los angeles', 'la', 'la county'],
        # México
        'cdmx': ['cdmx', 'ciudad de méxico', 'mexico df', 'distrito federal', 'azcapotzalco'],
        'puebla': ['puebla'],
        'aguascalientes': ['aguascalientes', 'ags'],
        # Canadá
        'canada federal': ['canada', 'federal', 'canadá', 'public land bank', 'canada public land bank'],
        'ontario': ['ontario', 'edwardsburgh', 'edwardsburgh-cardinal'],
        'regina': ['regina', 'saskatchewan']
    }

    for key, synonyms in sinonimos.items():
        if any(syn in consulta_lower for syn in synonyms):
            return LAND_BANKS.get(key)

    # Si no encuentra, buscar por país
    if 'mexico' in consulta_lower or 'méxico' in consulta_lower:
        return {'mensaje': '🇲🇽 LAND BANKS EN MÉXICO', 'tipo': 'pais', 'ciudades': ['cdmx', 'puebla', 'aguascalientes']}
    elif 'canada' in consulta_lower or 'canadá' in consulta_lower:
        return {'mensaje': '🇨🇦 LAND BANKS EN CANADÁ', 'tipo': 'pais', 'ciudades': ['canada federal', 'ontario', 'regina']}
    elif 'usa' in consulta_lower or 'estados unidos' in consulta_lower:
        return {'mensaje': '🇺🇸 LAND BANKS EN ESTADOS UNIDOS', 'tipo': 'pais', 'ciudades': ['tucson', 'phoenix', 'las vegas', 'los angeles']}

    return None

# ============================================
# FUNCIONES DE JARVIS
# ============================================

notas_guardadas = []

def formatear_land_bank(land_bank):
    """Formatea la información del land bank"""
    if not land_bank:
        return "❌ No encontré información para esa búsqueda"

    # Si es un país (varios land banks)
    if isinstance(land_bank, dict) and land_bank.get('tipo') == 'pais':
        respuesta = f"🏦 **{land_bank['mensaje']}**\n\n"
        for ciudad_key in land_bank['ciudades']:
            lb = LAND_BANKS.get(ciudad_key)
            if lb:
                respuesta += f"📍 **{lb['ubicacion']}** - {lb['nombre']}\n"
                respuesta += f"   🌐 {lb['sitio_web']}\n"
                respuesta += f"   🔗 Propiedades: {lb['propiedades_url']}\n\n"
        return respuesta

    # Land bank individual
    respuesta = f"""
    🏦 **{land_bank['nombre'].upper()}**
    {land_bank['pais']}
    📍 **Ubicación:** {land_bank['ubicacion']}

    📝 **Descripción:**
    {land_bank['descripcion']}

    🌐 **Sitio web:** {land_bank['sitio_web']}
    🔗 **Propiedades disponibles:** {land_bank['propiedades_url']}

    📞 **Contacto:** {land_bank['contacto']}

    🏠 **PROGRAMAS DISPONIBLES:**
    """

    for programa in land_bank['programas']:
        respuesta += f"   {programa}\n"

    if 'target_areas' in land_bank:
        respuesta += f"\n   🎯 **ÁREAS TARGET:**\n"
        for area in land_bank['target_areas']:
            respuesta += f"      • {area}\n"

    return respuesta

def procesar_comando(comando):
    """Procesa comandos del usuario"""
    comando = comando.lower().strip()

    if not comando:
        return "❌ No escribiste nada"

    # ===== LAND BANKS =====
    if 'land bank' in comando or 'landbank' in comando or 'banco de tierras' in comando:
        consulta = comando
        for palabra in ['land bank', 'landbank', 'banco de tierras', 'de', 'en', 'buscar', 'busca', 'ver', 'propiedades', 'del']:
            consulta = consulta.replace(palabra, '')
        consulta = consulta.strip()

        if not consulta or consulta == 'land':
            return """
            🏦 **LAND BANKS DISPONIBLES:**

            🇺🇸 **ESTADOS UNIDOS:**
            • 'land bank Tucson' - Southern Arizona Land Trust
            • 'land bank Phoenix' - Maricopa County
            • 'land bank Las Vegas' - Clark County
            • 'land bank Los Angeles' - LA County

            🇲🇽 **MÉXICO:**
            • 'land bank CDMX' - Banco de Tierras CDMX
            • 'land bank Puebla' - Banco de Tierras Puebla
            • 'land bank Aguascalientes' - Banco de Tierras Aguascalientes

            🇨🇦 **CANADÁ:**
            • 'land bank Canada' - Canada Public Land Bank
            • 'land bank Ontario' - Edwardsburgh Landbank
            • 'land bank Regina' - Regina Land Bank

            🌎 **PAÍSES:**
            • 'land bank México' - Ver todos en México
            • 'land bank Canadá' - Ver todos en Canadá
            • 'land bank USA' - Ver todos en USA
            """

        resultado = buscar_land_bank(consulta)
        if resultado:
            return formatear_land_bank(resultado)
        else:
            return f"❌ No encontré '{consulta}'. Prueba con una ciudad o país válido"

    # ===== VER TODOS =====
    elif 'todos los land banks' in comando or 'todos land banks' in comando:
        respuesta = "🏦 **LAND BANKS DISPONIBLES:**\n\n"
        for key, lb in LAND_BANKS.items():
            respuesta += f"{lb['pais']} **{lb['ubicacion']}** - {lb['nombre']}\n"
            respuesta += f"   Sitio: {lb['sitio_web']}\n\n"
        return respuesta

    # ===== ABRIR LAND BANK =====
    elif 'abrir' in comando and ('land bank' in comando or 'landbank' in comando):
        consulta = comando.replace('abrir', '').replace('land bank', '').replace('landbank', '').strip()
        if consulta:
            resultado = buscar_land_bank(consulta)
            if resultado and isinstance(resultado, dict) and 'propiedades_url' in resultado:
                display(HTML(f'<a href="{resultado["propiedades_url"]}" target="_blank">🏦 Abriendo {resultado["nombre"]}</a>'))
                return f"✅ Abriendo {resultado['nombre']} en nueva pestaña"
            else:
                return f"❌ No encontré '{consulta}' para abrir"
        else:
            return "❓ ¿Qué land bank quieres abrir? Ejemplo: 'abrir Las Vegas land bank'"

    # ===== PROPIEDADES NORMALES =====
    elif any(palabra in comando for palabra in ['casa', 'casas', 'propiedad', 'propiedades']):
        ubicacion = comando
        for palabra in ['buscar', 'casa', 'casas', 'propiedad', 'propiedades', 'en', 'venta']:
            ubicacion = ubicacion.replace(palabra, '')
        ubicacion = ubicacion.strip()

        if not ubicacion:
            return "❓ ¿En qué ciudad quieres buscar propiedades? Ejemplo: 'casas en Miami'"

        zillow_url = f"https://www.zillow.com/{urllib.parse.quote(ubicacion)}/for_sale/"
        display(HTML(f'<a href="{zillow_url}" target="_blank">🏠 Ver propiedades en Zillow</a>'))

        respuesta = f"🏠 **PROPIEDADES EN {ubicacion.upper()}**\n"
        respuesta += f"🔗 Ver en Zillow: [Haz clic aquí]({zillow_url})\n\n"

        # Verificar si hay land bank
        land_bank = buscar_land_bank(ubicacion)
        if land_bank and isinstance(land_bank, dict) and 'nombre' in land_bank:
            respuesta += f"💡 **¿Sabías que {ubicacion.title()} tiene un LAND BANK?**\n"
            respuesta += f"   Escribe 'land bank {ubicacion}' para ver propiedades recuperadas"

        return respuesta

    # ===== PAÍSES =====
    elif 'mexico' in comando or 'méxico' in comando:
        resultado = buscar_land_bank('mexico')
        if resultado:
            return formatear_land_bank(resultado)

    elif 'canada' in comando or 'canadá' in comando:
        resultado = buscar_land_bank('canada')
        if resultado:
            return formatear_land_bank(resultado)

    elif 'usa' in comando:
        resultado = buscar_land_bank('usa')
        if resultado:
            return formatear_land_bank(resultado)

    # ===== HORA Y FECHA =====
    elif 'hora' in comando:
        ahora = datetime.datetime.now().strftime("%I:%M %p")
        return f"🕐 Son las {ahora}"

    elif 'fecha' in comando:
        hoy = datetime.datetime.now().strftime("%d/%m/%Y")
        return f"📅 Hoy es {hoy}"

    # ===== NOTAS =====
    elif 'nota' in comando and 'agregar' in comando:
        nota = comando.replace('agregar', '').replace('nota', '').strip()
        if nota:
            notas_guardadas.append(nota)
            return f"📝 Nota guardada: {nota}"
        else:
            return "❓ Ejemplo: 'agregar nota revisar land bank Canadá'"

    elif 'mostrar notas' in comando:
        if notas_guardadas:
            notas_texto = "\n".join([f"  {i+1}. {n}" for i, n in enumerate(notas_guardadas)])
            return f"📋 Tus notas:\n{notas_texto}"
        else:
            return "📭 No hay notas guardadas"

    # ===== AYUDA =====
    elif 'ayuda' in comando:
        return """
        📢 **JARVIS - LAND BANKS USA, MÉXICO Y CANADÁ**
        ─────────────────────────────────────────────────────────────
        🌎 **COMANDOS POR PAÍS:**
        • 'land bank México' - Todos en México
        • 'land bank Canadá' - Todos en Canadá
        • 'land bank USA' - Todos en USA

        🇺🇸 **ESTADOS UNIDOS:**
        • 'land bank Tucson' - Southern Arizona
        • 'land bank Phoenix' - Maricopa County
        • 'land bank Las Vegas' - Clark County
        • 'land bank Los Angeles' - LA County

        🇲🇽 **MÉXICO:**
        • 'land bank CDMX' - Ciudad de México
        • 'land bank Puebla' - Banco de Tierras
        • 'land bank Aguascalientes' - Banco de Tierras

        🇨🇦 **CANADÁ:**
        • 'land bank Canada' - Federal Public Land Bank (90 propiedades)
        • 'land bank Ontario' - Edwardsburgh Landbank (10,000 acres)
        • 'land bank Regina' - Regina Land Bank (16 propiedades)

        🔗 **ABRIR:**
        • 'abrir [ciudad] land bank' - Abre el sitio web

        📝 **OTROS:**
        • 'hora' / 'fecha'
        • 'agregar nota [texto]' - Guardar nota
        • 'mostrar notas' - Ver notas
        • 'ayuda' - Este menú
        • 'salir' - Cerrar
        ─────────────────────────────────────────────────────────────
        """

    elif 'salir' in comando or 'adios' in comando:
        return "👋 ¡Hasta luego!"

    else:
        return f"❓ No entendí. Escribe 'ayuda' para comandos de land banks"


# ============================================
# INTERFAZ DE CHAT
# ============================================

input_box = widgets.Textarea(
    placeholder='Escribe tu comando aquí...\n\n📌 EJEMPLOS:\n• land bank Las Vegas\n• land bank CDMX\n• land bank Canadá\n• land bank México\n• abrir Canada land bank\n• ayuda',
    layout=widgets.Layout(width='100%', height='140px'),
    style={'description_width': 'initial'}
)

send_button = widgets.Button(
    description='📤 ENVIAR',
    button_style='primary',
    layout=widgets.Layout(width='auto')
)

clear_button = widgets.Button(
    description='🗑️ LIMPIAR CHAT',
    button_style='danger',
    layout=widgets.Layout(width='auto')
)

chat_output = widgets.Output()

def on_send_click(b):
    comando = input_box.value
    if comando.strip():
        with chat_output:
            print(f"\n{'─'*65}")
            print(f"👤 TÚ: {comando}")

            respuesta = procesar_comando(comando)
            print(f"🤖 JARVIS: {respuesta}")

        input_box.value = ''

        if 'salir' in comando.lower():
            send_button.disabled = True
            with chat_output:
                print("\n💡 Vuelve a ejecutar la celda para reiniciar")

def on_clear_click(b):
    with chat_output:
        clear_output(wait=True)
        print("✅ Chat limpiado")
        print("="*65)
        print("🌎 JARVIS - LAND BANKS USA, MÉXICO Y CANADÁ")
        print("📌 Prueba: 'land bank Canadá' o 'land bank CDMX'")
        print("📌 Todos: 'todos los land banks'")
        print("="*65)

send_button.on_click(on_send_click)
clear_button.on_click(on_clear_click)

display(widgets.VBox([
    widgets.HTML("<h3 style='text-align:center'>🌎 JARVIS - LAND BANKS EN USA, MÉXICO Y CANADÁ</h3>"),
    widgets.HTML("<p style='text-align:center; color:#666'>Propiedades gubernamentales en 🇺🇸 🇲🇽 🇨🇦</p>"),
    widgets.HBox([send_button, clear_button]),
    input_box,
    widgets.HTML("<b>💬 CHAT:</b>"),
    chat_output
]))

with chat_output:
    print("✅ Jarvis con Land Banks Internacionales está listo")
    print("="*65)
    print("🌎 **PRUEBA ESTOS COMANDOS:**")
    print("   🇺🇸 'land bank Las Vegas' - Clark County, NV")
    print("   🇲🇽 'land bank CDMX' - Banco de Tierras CDMX")
    print("   🇨🇦 'land bank Canadá' - Canada Public Land Bank (90 propiedades)")
    print("   🌎 'todos los land banks' - Ver todos")
    print("="*65)
```
