```python
# ============================================
# JARVIS CON LAND BANKS - SIN INSTALAR NADA
# Funciona en cualquier Colab sin pip
# ============================================

import datetime
import random
import urllib.parse
import urllib.request
import json
from IPython.display import display, HTML, clear_output
import ipywidgets as widgets

clear_output(wait=True)

print("="*60)
print("🏦 JARVIS - LAND BANKS USA (SIN INSTALACIONES)")
print("="*60)
print("✅ Funciona sin instalar nada - usa urllib nativo")
print("="*60)

# ============================================
# BASE DE DATOS DE LAND BANKS
# ============================================

LAND_BANKS = {
    'tucson': {
        'nombre': 'Southern Arizona Land Trust (SALT)',
        'ubicacion': 'Tucson, Arizona',
        'county': 'Pima County',
        'descripcion': 'Organización sin fines de lucro que revitaliza áreas afectadas por la recesión, con 128 propiedades en alquiler y múltiples proyectos en desarrollo.',
        'sitio_web': 'https://www.saltaz.org',
        'propiedades_url': 'https://www.saltaz.org/properties',
        'programas': [
            '🏠 Propiedades en alquiler - 128 unidades (20% para familias de bajos ingresos)',
            '🏗️ Belnor Vista - Desarrollo de viviendas (planes de servicios públicos en marcha)',
            '🏘️ Bentley on the Loop - Mejoras completadas para 3 nuevas casas',
            '🔨 Emery Park Place - Preparación para 4 dúplex',
            '📐 Pinal Vista Place - División en 6 parcelas para viviendas unifamiliares'
        ],
        'contacto': '3044 N Alvernon Way, Tucson, AZ 85712',
        'tipo': 'land_trust'
    },

    'phoenix': {
        'nombre': 'Maricopa County Land Bank / City of Phoenix',
        'ubicacion': 'Phoenix, Arizona',
        'county': 'Maricopa County',
        'descripcion': 'Programas de estabilización de vecindarios y propiedades en ejecución fiscal del condado de Maricopa.',
        'sitio_web': 'https://www.maricopa.gov/5647/Land-Bank',
        'propiedades_url': 'https://www.maricopa.gov/5651/Available-Properties',
        'programas': [
            '🏠 Neighborhood Stabilization Program - Viviendas recuperadas',
            '📋 Tax Lien Sales - Propiedades con deuda fiscal',
            '🔨 Land Reutilization - Terrenos para reurbanización',
            '🏘️ Affordable Housing Program - Vivienda asequible'
        ],
        'contacto': '301 W Jefferson St, Phoenix, AZ 85003',
        'tipo': 'county'
    },

    'las vegas': {
        'nombre': 'Clark County Land Bank (Land Reutilization Corporation)',
        'ubicacion': 'Las Vegas, Nevada',
        'county': 'Clark County',
        'descripcion': 'Programa establecido para eliminar deterioro urbano y estabilizar vecindarios mediante la adquisición y reurbanización de propiedades problemáticas.',
        'sitio_web': 'https://clarkcountylandbank.org',
        'propiedades_url': 'https://clarkcountylandbank.org/available-properties',
        'programas': [
            '🏠 Side Lot Program - Adquiere terrenos vacantes adyacentes a tu propiedad',
            '🏗️ Vacant Lot Program - Compra terrenos NO adyacentes para construir viviendas',
            '🏢 Commercial Property - Adquisición para desarrollo económico',
            '🔨 Rehabilitation Program - Compra y renovación con requisito de ocupación de 3 años',
            '🏚️ Demolition Program - Eliminación de estructuras deterioradas (132 demoliciones completadas)'
        ],
        'target_areas': [
            'Grand Ave South / Promise Neighborhood',
            'Highlands / Southgate Neighborhood',
            'Lagonda Redevelopment Corridor',
            'Sheridan / Kenton Corridor',
            'South Yellow Springs Corridor',
            'Western Edge'
        ],
        'contacto': '500 S Grand Central Pkwy, Las Vegas, NV 89106',
        'tipo': 'county'
    },

    'los angeles': {
        'nombre': 'Los Angeles County Land Bank Pilot Program',
        'ubicacion': 'Los Angeles, California',
        'county': 'Los Angeles County',
        'descripcion': 'Programa piloto lanzado en 2022 con $50 millones para prevenir desplazamiento cerca de nuevas infraestructuras (Metro y LA River).',
        'sitio_web': 'https://ceo.lacounty.gov/land-bank',
        'propiedades_url': 'https://ceo.lacounty.gov/land-bank/available-properties',
        'programas': [
            '🏗️ Adquisición cerca de estaciones de Metro para vivienda asequible',
            '🏠 2615 W. 8th Street - 27,641 sq ft comprado por $4.3M para 160 unidades residenciales + retail',
            '🚆 Propiedad cerca de A Line en Downtown Long Beach - 160 viviendas + usos comerciales',
            '🤝 Colaboración con LA Community Land Trust Coalition',
            '🚇 Coordinación con Metro para propiedades cerca de tránsito'
        ],
        'contacto': '500 W Temple St, Los Angeles, CA 90012',
        'tipo': 'pilot'
    }
}

# ============================================
# FUNCIONES DE BÚSQUEDA
# ============================================

def buscar_land_bank(ciudad):
    """Busca información de land bank por ciudad"""
    ciudad_lower = ciudad.lower()

    sinonimos = {
        'tucson': ['tucson', 'southern arizona', 'salt', 'pima county'],
        'phoenix': ['phoenix', 'maricopa', 'maricopa county'],
        'las vegas': ['las vegas', 'vegas', 'clark county', 'clark'],
        'los angeles': ['los angeles', 'la', 'la county', 'los angeles county']
    }

    for key, synonyms in sinonimos.items():
        if any(syn in ciudad_lower for syn in synonyms):
            return LAND_BANKS[key]

    return None

# ============================================
# FUNCIONES DE JARVIS
# ============================================

notas_guardadas = []

def formatear_land_bank(land_bank):
    """Formatea la información del land bank"""
    respuesta = f"""
    🏦 **{land_bank['nombre'].upper()}**
    📍 **Ubicación:** {land_bank['ubicacion']}
    📋 **Condado:** {land_bank['county']}

    📝 **Descripción:**
    {land_bank['descripcion']}

    🌐 **Sitio web:** {land_bank['sitio_web']}
    🔗 **Propiedades disponibles:** {land_bank['propiedades_url']}

    📞 **Contacto:** {land_bank['contacto']}

    🏠 **PROGRAMAS DISPONIBLES:**
    """

    for programa in land_bank['programas']:
        respuesta += f"   {programa}\n"

    if 'target_areas' in land_bank:
        respuesta += f"\n   🎯 **ÁREAS TARGET:**\n"
        for area in land_bank['target_areas']:
            respuesta += f"      • {area}\n"

    return respuesta

def procesar_comando(comando):
    """Procesa comandos del usuario"""
    comando = comando.lower().strip()

    if not comando:
        return "❌ No escribiste nada"

    # ===== LAND BANKS =====
    if 'land bank' in comando or 'landbank' in comando:
        ciudad = comando
        for palabra in ['land bank', 'landbank', 'de', 'en', 'buscar', 'busca', 'ver', 'propiedades', 'del']:
            ciudad = ciudad.replace(palabra, '')
        ciudad = ciudad.strip()

        if not ciudad or ciudad == 'land':
            return """
            🏦 **LAND BANKS DISPONIBLES:**

            • 'land bank Tucson' - Southern Arizona Land Trust
            • 'land bank Phoenix' - Maricopa County
            • 'land bank Las Vegas' - Clark County Land Bank
            • 'land bank Los Angeles' - LA County Land Bank
            • 'todos los land banks' - Ver todos
            """

        land_bank = buscar_land_bank(ciudad)
        if land_bank:
            return formatear_land_bank(land_bank)
        else:
            return f"❌ No encontré '{ciudad}'. Prueba: Tucson, Phoenix, Las Vegas, Los Angeles"

    # ===== VER TODOS =====
    elif 'todos los land banks' in comando:
        respuesta = "🏦 **LAND BANKS DISPONIBLES:**\n\n"
        for key, lb in LAND_BANKS.items():
            respuesta += f"📍 **{lb['ubicacion']}** - {lb['nombre']}\n"
            respuesta += f"   Sitio: {lb['sitio_web']}\n"
            respuesta += f"   Propiedades: {lb['propiedades_url']}\n\n"
        return respuesta

    # ===== ABRIR LAND BANK =====
    elif 'abrir' in comando and ('land bank' in comando or 'landbank' in comando):
        ciudad = comando.replace('abrir', '').replace('land bank', '').replace('landbank', '').strip()
        if ciudad:
            land_bank = buscar_land_bank(ciudad)
            if land_bank:
                display(HTML(f'<a href="{land_bank["propiedades_url"]}" target="_blank">🏦 Abriendo {land_bank["nombre"]}</a>'))
                return f"✅ Abriendo {land_bank['nombre']} en nueva pestaña"
            else:
                return f"❌ No encontré '{ciudad}'"
        else:
            return "❓ ¿Qué land bank quieres abrir? Ejemplo: 'abrir Las Vegas land bank'"

    # ===== PROPIEDADES NORMALES =====
    elif any(palabra in comando for palabra in ['casa', 'casas', 'propiedad', 'propiedades']):
        ubicacion = comando
        for palabra in ['buscar', 'casa', 'casas', 'propiedad', 'propiedades', 'en', 'venta']:
            ubicacion = ubicacion.replace(palabra, '')
        ubicacion = ubicacion.strip()

        if not ubicacion:
            return "❓ ¿En qué ciudad quieres buscar propiedades? Ejemplo: 'casas en Miami'"

        land_bank = buscar_land_bank(ubicacion)
        zillow_url = f"https://www.zillow.com/{urllib.parse.quote(ubicacion)}/for_sale/"
        display(HTML(f'<a href="{zillow_url}" target="_blank">🏠 Ver propiedades en Zillow</a>'))

        respuesta = f"🏠 **PROPIEDADES EN {ubicacion.upper()}**\n"
        respuesta += f"🔗 Ver en Zillow: [Haz clic aquí]({zillow_url})\n\n"

        if land_bank:
            respuesta += f"💡 **¿Sabías que {ubicacion.title()} tiene un LAND BANK?**\n"
            respuesta += f"   Escribe 'land bank {ubicacion}' para ver propiedades recuperadas"

        return respuesta

    # ===== HORA Y FECHA =====
    elif 'hora' in comando:
        ahora = datetime.datetime.now().strftime("%I:%M %p")
        return f"🕐 Son las {ahora}"

    elif 'fecha' in comando:
        hoy = datetime.datetime.now().strftime("%d/%m/%Y")
        return f"📅 Hoy es {hoy}"

    # ===== NOTAS =====
    elif 'nota' in comando and 'agregar' in comando:
        nota = comando.replace('agregar', '').replace('nota', '').strip()
        if nota:
            notas_guardadas.append(nota)
            return f"📝 Nota guardada: {nota}"
        else:
            return "❓ Ejemplo: 'agregar nota revisar land bank Las Vegas'"

    elif 'mostrar notas' in comando:
        if notas_guardadas:
            notas_texto = "\n".join([f"  {i+1}. {n}" for i, n in enumerate(notas_guardadas)])
            return f"📋 Tus notas:\n{notas_texto}"
        else:
            return "📭 No hay notas guardadas"

    # ===== AYUDA =====
    elif 'ayuda' in comando:
        return """
        📢 **JARVIS - LAND BANKS USA**
        ─────────────────────────────────────────────
        🏦 **LAND BANKS:**
        • "land bank Tucson" - Southern Arizona
        • "land bank Phoenix" - Maricopa County
        • "land bank Las Vegas" - Clark County
        • "land bank Los Angeles" - LA County
        • "todos los land banks" - Ver todos

        🔗 **ABRIR:**
        • "abrir [ciudad] land bank" - Abre el sitio

        🏠 **PROPIEDADES:**
        • "casas en Miami" - Ver en Zillow

        📝 **NOTAS:**
        • "agregar nota [texto]" - Guardar nota
        • "mostrar notas" - Ver notas

        ⏰ **OTROS:**
        • "hora" / "fecha"
        • "ayuda" - Este menú
        • "salir" - Cerrar
        ─────────────────────────────────────────────
        """

    elif 'salir' in comando or 'adios' in comando:
        return "👋 ¡Hasta luego!"

    else:
        return f"❓ No entendí. Escribe 'ayuda' para comandos de land banks"


# ============================================
# INTERFAZ DE CHAT
# ============================================

input_box = widgets.Textarea(
    placeholder='Escribe tu comando aquí...\n\n📌 EJEMPLOS:\n• land bank Las Vegas\n• land bank Tucson\n• land bank Los Angeles\n• abrir Phoenix land bank\n• todos los land banks\n• casas en Miami\n• ayuda',
    layout=widgets.Layout(width='100%', height='140px'),
    style={'description_width': 'initial'}
)

send_button = widgets.Button(
    description='📤 ENVIAR',
    button_style='primary',
    layout=widgets.Layout(width='auto')
)

clear_button = widgets.Button(
    description='🗑️ LIMPIAR CHAT',
    button_style='danger',
    layout=widgets.Layout(width='auto')
)

chat_output = widgets.Output()

def on_send_click(b):
    comando = input_box.value
    if comando.strip():
        with chat_output:
            print(f"\n{'─'*60}")
            print(f"👤 TÚ: {comando}")

            respuesta = procesar_comando(comando)
            print(f"🤖 JARVIS: {respuesta}")

        input_box.value = ''

        if 'salir' in comando.lower():
            send_button.disabled = True
            with chat_output:
                print("\n💡 Vuelve a ejecutar la celda para reiniciar")

def on_clear_click(b):
    with chat_output:
        clear_output(wait=True)
        print("✅ Chat limpiado")
        print("="*60)
        print("🏦 JARVIS - LAND BANKS LISTO")
        print("📌 Prueba: 'land bank Las Vegas'")
        print("📌 Todos: 'todos los land banks'")
        print("="*60)

send_button.on_click(on_send_click)
clear_button.on_click(on_clear_click)

display(widgets.VBox([
    widgets.HTML("<h3 style='text-align:center'>🏦 JARVIS - LAND BANKS EN USA</h3>"),
    widgets.HTML("<p style='text-align:center; color:#666'>Propiedades de land banks en Tucson, Phoenix, Las Vegas y Los Angeles</p>"),
    widgets.HBox([send_button, clear_button]),
    input_box,
    widgets.HTML("<b>💬 CHAT:</b>"),
    chat_output
]))

with chat_output:
    print("✅ Jarvis con Land Banks está listo")
    print("="*60)
    print("🏦 **PRUEBA ESTOS COMANDOS:**")
    print("   • 'land bank Las Vegas' - Clark County")
    print("   • 'land bank Los Angeles' - LA County")
    print("   • 'land bank Tucson' - Southern Arizona")
    print("   • 'land bank Phoenix' - Maricopa County")
    print("   • 'todos los land banks' - Ver todos")
    print("="*60)
```
